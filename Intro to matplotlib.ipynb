```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
```

```python
# showing the years and runs data into a graph using matplotlib
years = [1990, 1991, 1992, 1994, 1996]
runs = [500, 200, 1000, 1200, 4000]

plt.plot(years , runs)
```

```python
# Adding the x and y labels
plt.plot(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin's Yearly runs")
```

```python
# multiple lines in one plot image
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years, kohli, label = "Virat kohli")
plt.plot(years, sehwag, label = "Virender sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance comparision")
plt.legend()
plt.show()
```

```python
# Multi lines in one plot
years = [2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend(loc = "upper left")
plt.show()
```

```python
# using format string
plt.plot(years, kohli, 'ro--', label = "kohli") # red circles with dashed lines
plt.plot(years, sehwag, 'g^:', label = "sehwag")
plt.legend()
```

```python
# color and line style arguments
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color = "purple", linestyle= 'dashed', label = "Sehwag")
plt.legend()
```
